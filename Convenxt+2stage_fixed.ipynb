```python
# ğŸ“Œ [4] dataload.py (ResizePadToSquare ë¨¼ì €â†’PIL ê¸°ë°˜)
import random, math, numpy as np, torch
from torch.utils.data import Dataset, DataLoader, Subset, WeightedRandomSampler
from torchvision.transforms import v2, functional as F
from sklearn.model_selection import train_test_split
import zipfile, shutil, os
from PIL import Image
with open('/content/config.yaml') as f:
    CFG = yaml.safe_load(f)

def seed_everything(s):
    random.seed(s); np.random.seed(s)
    torch.manual_seed(s); torch.cuda.manual_seed_all(s)

class ResizePadToSquare(torch.nn.Module):
    def __init__(self, size:int, fill=114):
        super().__init__(); self.size, self.fill = size, fill
    def forward(self, img:Image.Image):
        w, h = img.size
        scale = self.size / max(w, h)
        img = F.resize(img, (int(h*scale), int(w*scale)), antialias=True)
        new_w, new_h = img.size
        pad_left  = (self.size - new_w) // 2
        pad_top   = (self.size - new_h) // 2
        pad_right = self.size - new_w - pad_left
        pad_bot   = self.size - new_h - pad_top
        return F.pad(img, [pad_left, pad_top, pad_right, pad_bot], fill=self.fill)

def _transforms():
    pad = ResizePadToSquare(CFG['IMG_SIZE'])
    train_tf = v2.Compose([
        pad,
        v2.ToImage(),
        v2.ToDtype(torch.float32, scale=True),
        v2.RandomHorizontalFlip(), v2.RandomVerticalFlip(0.3),
        v2.RandomRotation(15, fill=114),
        v2.ColorJitter(0.2,0.2,0.2),
        v2.GaussianBlur(3,(0.1,2.0)),
        v2.RandomAutocontrast(), v2.RandomErasing(p=0.25),
        v2.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
    ])
    val_tf = v2.Compose([
        pad,
        v2.ToImage(),
        v2.ToDtype(torch.float32, scale=True),
        v2.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
    ])
    return train_tf, val_tf

class ImageDS(Dataset):
    def __init__(self, root, tf=None, test=False):
        self.tf, self.test = tf, test; self.samples=[]
        if test:
            self.samples = [os.path.join(root,f) for f in sorted(os.listdir(root)) if f.lower().endswith(('jpg','png'))]
        else:
            for cls in sorted(os.listdir(root)):
                for f in os.listdir(os.path.join(root,cls)):
                    p=os.path.join(root,cls,f)
                    if p.lower().endswith(('jpg','png')):
                        self.samples.append((p, cls))
            self.class_to_idx={c:i for i,c in enumerate(sorted({c for _,c in self.samples}))}
            self.samples=[(p,self.class_to_idx[c]) for p,c in self.samples]
    def __len__(self): return len(self.samples)
    def __getitem__(self, idx):
        if self.test:
            img=Image.open(self.samples[idx]).convert('RGB')
            return (self.tf(img) if self.tf else img, self.samples[idx])
        path,label = self.samples[idx]; img=Image.open(path).convert('RGB')
        return (self.tf(img) if self.tf else img, label)

def get_loaders(train_root='/content/open_utf/open/train', test_root='/content/open_utf/open/test'):
    seed_everything(CFG['SEED'])
    full=ImageDS(train_root)
    y=[l for _,l in full.samples]
    tr_idx,val_idx = train_test_split(range(len(full)), test_size=0.2, stratify=y, random_state=CFG['SEED'])
    tr_tf,val_tf = _transforms()
    tr_ds = Subset(ImageDS(train_root,tr_tf), tr_idx)
    val_ds= Subset(ImageDS(train_root,val_tf), val_idx)
    counts=np.bincount(y); weights=[1.0/counts[l] for l in y]
    sampler=WeightedRandomSampler([weights[i] for i in tr_idx], len(tr_ds), True)
    dl_tr = DataLoader(tr_ds,batch_size=CFG['BATCH_SIZE'],sampler=sampler,num_workers=4,pin_memory=True)
    dl_val= DataLoader(val_ds,batch_size=CFG['BATCH_SIZE'],shuffle=False,num_workers=2,pin_memory=True)
    dl_test=DataLoader(ImageDS(test_root,val_tf,True),batch_size=CFG['BATCH_SIZE'],shuffle=False,num_workers=2)
    class_names=sorted(full.class_to_idx,key=full.class_to_idx.get)
    return dl_tr, dl_val, dl_test, class_names, counts
```

```python
import os, gc, random, math, yaml
import numpy as np
from pathlib import Path
from PIL import Image
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, Subset, WeightedRandomSampler
from torch.cuda.amp import GradScaler, autocast
from sklearn.model_selection import StratifiedKFold

import timm
from transformers import get_cosine_schedule_with_warmup
from torchvision.transforms import functional as TF

# ------------------------- utility ------------------------- #
def seed_everything(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = False
    torch.backends.cudnn.benchmark = True

# ------------------------- dataset ------------------------ #
class ResizePadToSquare(nn.Module):
    def __init__(self, size: int, fill: int = 114):
        super().__init__()
        self.size, self.fill = size, fill

    def forward(self, img: Image.Image):
        w, h = img.size
        scale = self.size / max(w, h)
        img = TF.resize(img, (int(h * scale), int(w * scale)), antialias=True)
        new_w, new_h = img.size
        pad_left = (self.size - new_w) // 2
        pad_top = (self.size - new_h) // 2
        pad_right = self.size - new_w - pad_left
        pad_bot = self.size - new_h - pad_top
        return TF.pad(img, [pad_left, pad_top, pad_right, pad_bot], fill=self.fill)

def build_transforms(img_size: int):
    import torchvision.transforms as T
    pad = ResizePadToSquare(img_size)
    train_tf = T.Compose([
        pad,
        T.ToTensor(),
        T.RandomHorizontalFlip(),
        T.RandomVerticalFlip(p=0.3),
        T.RandomRotation(15, fill=0),
        T.ColorJitter(0.2, 0.2, 0.2),
        T.GaussianBlur(3, (0.1, 2.0)),
        T.RandomAutocontrast(),
        T.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
    ])
    val_tf = T.Compose([
        pad,
        T.ToTensor(),
        T.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
    ])
    return train_tf, val_tf

class RawImageDS(Dataset):
    EXT = ("jpg", "jpeg", "png", "bmp")

    def __init__(self, root: str):
        self.samples = []
        class_names = sorted([d for d in os.listdir(root) if os.path.isdir(os.path.join(root, d))])
        self.class_to_idx = {c: i for i, c in enumerate(class_names)}
        for cls in class_names:
            cls_dir = Path(root) / cls
            for fname in cls_dir.iterdir():
                if fname.suffix.lower().lstrip(".") in self.EXT:
                    self.samples.append((str(fname), self.class_to_idx[cls]))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]
        img = Image.open(path).convert("RGB")
        return img, label

class SoftLabelDataset(Dataset):
    def __init__(self, raw_ds: RawImageDS, softs: torch.Tensor, tf):
        self.raw_ds = raw_ds
        self.softs = softs
        self.tf = tf

    def __len__(self):
        return len(self.raw_ds)

    def __getitem__(self, idx):
        img, label = self.raw_ds[idx]
        return self.tf(img), label, self.softs[idx]

# ------------------------- model --------------------------- #
class CustomModel(nn.Module):
    def __init__(self, model_name: str, num_classes: int):
        super().__init__()
        self.backbone = timm.create_model(model_name, pretrained=True, num_classes=num_classes)

    def forward(self, x):
        return self.backbone(x)

def create_model(model_name: str, num_classes: int):
    return CustomModel(model_name, num_classes)

# ------------------------- KD training -------------------- #
def train_kd(cfg, teacher_path: str, train_root: str, save_dir: str):
    seed_everything(cfg['SEED'])
    os.makedirs(save_dir, exist_ok=True)  # âœ… ë””ë ‰í† ë¦¬ ìƒì„± ë³´ì¥
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    IMG_SIZE = cfg['IMG_SIZE']
    BATCH_SIZE = cfg['BATCH_SIZE']
    EPOCHS = cfg['EPOCHS']
    LR = cfg['LEARNING_RATE']
    PATIENCE = cfg['patience']
    TEMPERATURE = 2.0
    ALPHA = 0.7

    raw_ds = RawImageDS(train_root)
    labels = [l for _, l in raw_ds.samples]
    num_classes = len(set(labels))

    cache_path = Path(save_dir) / "soft_labels.pt"
    if cache_path.exists():
        print("ğŸ”„ soft_labels.pt ë¡œë“œ ì¤‘â€¦")
        soft_labels = torch.load(cache_path, map_location='cpu')
    else:
        print("ğŸ§  Generating soft labels from Teacherâ€¦")
        val_tf = build_transforms(IMG_SIZE)[1]
        teacher_ds = SoftLabelDataset(raw_ds, torch.zeros(len(raw_ds), num_classes), val_tf)
        teacher_loader = DataLoader(teacher_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=4)

        teacher = create_model(cfg['model'], num_classes).to(device)
        teacher.load_state_dict(torch.load(teacher_path, map_location=device))
        teacher.eval();  [p.requires_grad_(False) for p in teacher.parameters()]

        soft_chunks = []
        with torch.no_grad():
            for imgs, _, _ in tqdm(teacher_loader, desc="Teacher Inference"):
                imgs = imgs.to(device)
                logits = teacher(imgs)
                soft = F.softmax(logits / TEMPERATURE, dim=1).cpu().half()
                soft_chunks.append(soft)
        soft_labels = torch.cat(soft_chunks)
        torch.save(soft_labels, cache_path)
        del teacher, soft_chunks; torch.cuda.empty_cache()

    train_tf, val_tf = build_transforms(IMG_SIZE)
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=cfg['SEED'])

    class_counts = np.bincount(labels)
    class_weights = torch.tensor(1.0 / class_counts, dtype=torch.float32)

    for fold, (tr_idx, va_idx) in enumerate(skf.split(np.zeros(len(labels)), labels), 1):
        if fold < 3:
          continue  # âœ… Fold 1, 2ëŠ” ê±´ë„ˆëœ€
        if fold > 4:
          continue
        print(f"\nğŸ” Fold {fold}/5")
        tr_ds = Subset(SoftLabelDataset(raw_ds, soft_labels, tf=train_tf), tr_idx)
        va_ds = Subset(SoftLabelDataset(raw_ds, soft_labels, tf=val_tf), va_idx)

        tr_weights = [class_weights[labels[i]] for i in tr_idx]
        sampler = WeightedRandomSampler(tr_weights, num_samples=len(tr_weights), replacement=True)

        tr_loader = DataLoader(tr_ds, batch_size=BATCH_SIZE, sampler=sampler, num_workers=4, pin_memory=True)
        va_loader = DataLoader(va_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=4)

        student = create_model(cfg['model'], num_classes).to(device)
        optimizer = torch.optim.AdamW(student.parameters(), lr=LR, weight_decay=1e-4)
        total_steps = len(tr_loader) * EPOCHS
        scheduler = get_cosine_schedule_with_warmup(optimizer, len(tr_loader)*2, total_steps)
        scaler = GradScaler()

        ce_crit = nn.CrossEntropyLoss(weight=class_weights.to(device))
        kd_crit = nn.KLDivLoss(reduction='batchmean')

        best_loss = float('inf'); patience_cnt = 0
        for epoch in range(1, EPOCHS+1):
            student.train();  running_loss = 0; correct = 0; n = 0
            pbar = tqdm(tr_loader, desc=f"Fold{fold} Epoch{epoch}")
            for imgs, lbls, softs in pbar:
                imgs, lbls, softs = imgs.to(device), lbls.to(device), softs.to(device)
                optimizer.zero_grad()
                with autocast(dtype=torch.float16):
                    logits = student(imgs)
                    kd_loss = kd_crit(F.log_softmax(logits / TEMPERATURE, dim=1), softs) * (TEMPERATURE**2)
                    ce_loss = ce_crit(logits, lbls)
                    loss = ALPHA*kd_loss + (1-ALPHA)*ce_loss
                scaler.scale(loss).backward()
                torch.nn.utils.clip_grad_norm_(student.parameters(), 5.0)
                scaler.step(optimizer)
                scaler.update(); scheduler.step()

                running_loss += loss.item() * imgs.size(0)
                n += imgs.size(0); correct += (logits.argmax(1)==lbls).sum().item()
                pbar.set_postfix(train_loss=running_loss/n, train_acc=correct/n)

            student.eval();  val_loss = 0; val_correct = 0; val_n = 0
            with torch.no_grad():
                for imgs, lbls, softs in va_loader:
                    imgs, lbls, softs = imgs.to(device), lbls.to(device), softs.to(device)
                    with autocast(dtype=torch.float16):
                        logits = student(imgs)
                        kd_loss = kd_crit(F.log_softmax(logits / TEMPERATURE, dim=1), softs) * (TEMPERATURE**2)
                        ce_loss = ce_crit(logits, lbls)
                        loss = ALPHA*kd_loss + (1-ALPHA)*ce_loss
                    val_loss += loss.item() * imgs.size(0)
                    val_correct += (logits.argmax(1)==lbls).sum().item()
                    val_n += lbls.size(0)
            val_loss /= val_n;  val_acc = val_correct/val_n
            print(f"ğŸ”µ Val Loss: {val_loss:.4f} | Acc: {val_acc:.4f}")

            if val_loss < best_loss:
                best_loss = val_loss; patience_cnt = 0
                ckpt_path = Path(save_dir) / f"Student_fold{fold}_best.pth"
                torch.save(student.state_dict(), ckpt_path)
                print(f"âœ… Saved: {ckpt_path.name}")
            else:
                patience_cnt += 1
                if patience_cnt >= PATIENCE:
                    print("ğŸ›‘ EarlyStop triggered."); break

        torch.cuda.empty_cache(); gc.collect()

# ------------------------- run in Colab ---------------------------- #
if __name__ == "__main__":
    CFG = yaml.safe_load(open("/content/config.yaml"))
    train_kd(CFG,
             teacher_path="/content/drive/MyDrive/YSG_Model/sg_convnext_B_Reclean(0.156).pth",
             train_root="/content/open_utf/open/train",
             save_dir="/content/drive/MyDrive/YSG_Model/Student_KD_Fold_3")

```

```python
import torch
import pandas as pd, os
from tqdm import tqdm

# â”€â”€â”€â”€â”€ ì„¤ì • â”€â”€â”€â”€â”€ #
FOLD_RANGE = range(1, 6)
MODEL_DIR = "/content/drive/MyDrive/YSG_Model/Student"
MODEL_NAME_FMT = "Student_fold{}_best.pth"
SAVE_PATH = "/content/drive/MyDrive/Final_Student_TTA_MAX.csv"
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# â”€â”€â”€â”€â”€ ë³µìˆ˜ ì •ë‹µ ëŒ€ì‘ í´ë˜ìŠ¤ ë§¤í•‘ â”€â”€â”€â”€â”€ #
EQUIV_CLASSES = {
    'K5_3ì„¸ëŒ€_í•˜ì´ë¸Œë¦¬ë“œ_2020_2022': 'K5_í•˜ì´ë¸Œë¦¬ë“œ_3ì„¸ëŒ€_2020_2023',
    'ë””_ì˜¬ë‰´ë‹ˆë¡œ_2022_2025': 'ë””_ì˜¬_ë‰´_ë‹ˆë¡œ_2022_2025',
    '718_ë°•ìŠ¤í„°_2017_2024': 'ë°•ìŠ¤í„°_718_2017_2024',
    'RAV4_2016_2018': 'ë¼ë¸Œ4_4ì„¸ëŒ€_2013_2018',
    'RAV4_5ì„¸ëŒ€_2019_2024': 'ë¼ë¸Œ4_5ì„¸ëŒ€_2019_2024',
}

def apply_equiv_class_postprocess(pred_df, class_names):
    class_map = {cls: EQUIV_CLASSES.get(cls, cls) for cls in class_names}
    new_records = []
    for _, row in pred_df.iterrows():
        id_ = row["ID"]
        probs = row[class_names].values
        merged = {}
        for i, cls in enumerate(class_names):
            rep = class_map[cls]
            merged[rep] = max(merged.get(rep, 0), probs[i])
        final = {cls: merged.get(cls, 0) for cls in class_names}
        final["ID"] = id_
        new_records.append(final)
    return pd.DataFrame(new_records)[["ID"] + class_names]

# â”€â”€â”€â”€â”€ ëª¨ë¸ ë¡œë”© â”€â”€â”€â”€â”€ #
model_list = []
for fold in FOLD_RANGE:
    path = os.path.join(MODEL_DIR, MODEL_NAME_FMT.format(fold))
    model = Net(len(class_names))  # âœ… class_namesëŠ” ì™¸ë¶€ ì •ì˜
    model.load_state_dict(torch.load(path, map_location=DEVICE))
    model.to(DEVICE)
    model.eval()
    model_list.append(model)

# â”€â”€â”€â”€â”€ TTA ì •ì˜ â”€â”€â”€â”€â”€ #
def apply_tta(imgs):
    return [imgs, torch.flip(imgs, dims=[3])]  # ì›ë³¸ + ì¢Œìš° ë°˜ì „

# â”€â”€â”€â”€â”€ DataLoader â”€â”€â”€â”€â”€ #
_, _, test_dl, _, _ = get_loaders()

# â”€â”€â”€â”€â”€ ì¶”ë¡  (Fold+TTA í™•ë¥  MAX) â”€â”€â”€â”€â”€ #
records = []
with torch.no_grad():
    for imgs, paths in tqdm(test_dl, desc="ğŸ² Inference (Fold+TTA Max Strategy)"):
        imgs = imgs.to(DEVICE)
        fold_preds = []
        for model in model_list:
            tta_probs = [model(aug).softmax(1) for aug in apply_tta(imgs)]
            fold_max = torch.stack(tta_probs).max(dim=0)[0]
            fold_preds.append(fold_max)
        final_probs = torch.stack(fold_preds).max(dim=0)[0].cpu()

        for pr, path in zip(final_probs, paths):
            rec = {cls: pr[i].item() for i, cls in enumerate(class_names)}
            rec["ID"] = os.path.splitext(os.path.basename(path))[0]
            records.append(rec)

# â”€â”€â”€â”€â”€ í›„ì²˜ë¦¬ ë° ì €ì¥ â”€â”€â”€â”€â”€ #
submission_df = pd.DataFrame(records)[["ID"] + class_names]
submission_df = apply_equiv_class_postprocess(submission_df, class_names)
submission_df.to_csv(SAVE_PATH, index=False, encoding="utf-8-sig")
print(f"âœ… ë„ë°• ì „ëµ ê²°ê³¼ ì €ì¥ ì™„ë£Œ âœ {SAVE_PATH}")
```
